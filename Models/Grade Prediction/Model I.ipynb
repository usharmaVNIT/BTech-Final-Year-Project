```python
img_width, img_height = 150, 150
batch_size = 32
epochs = 5


train_data_dir = '/content/drive/MyDrive/Final year project/CNN_test/Grade/train'
test_data_dir = '/content/drive/MyDrive/Final year project/CNN_test/Grade/test'
```

```python
from keras.applications import vgg16

#base_model = tf.keras.applications.MobileNetV2(input_shape=(150, 150, 3),include_top=False,weights='imagenet')

base_model = vgg16.VGG16(weights='imagenet', include_top=False, input_shape=(150, 150, 3))

for layer in base_model.layers:
    layer.trainable = False
```

```python
FC_Head = addTopOnNet(base_model, 3)

model = Model(inputs = base_model.input, outputs = FC_Head)

print(model.summary())
```

```python
train_datagen = ImageDataGenerator(rescale=1. / 255)

test_datagen = ImageDataGenerator(rescale=1. / 255)

train_generator = train_datagen.flow_from_directory(
    train_data_dir,
    target_size=(img_width, img_height),
    batch_size=batch_size,
    class_mode='categorical')

validation_generator = test_datagen.flow_from_directory(
    test_data_dir,
    target_size=(img_width, img_height),
    batch_size=batch_size,
    class_mode='categorical')
```

```python
nb_train_samples = 5757
nb_test_samples = 1646
```

```python
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy', tf.keras.metrics.Precision(), tf.keras.metrics.Recall()])
```

```python
model.fit_generator(
    train_generator,
    steps_per_epoch=nb_train_samples // batch_size,
    epochs=epochs,
    validation_data=train_generator,
    validation_steps=nb_test_samples // batch_size)
```
