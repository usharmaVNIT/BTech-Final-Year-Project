```python
def calc_area(x, y):
    return 0.5*np.abs(np.dot(x,np.roll(y,1))-np.dot(y,np.roll(x,1)))
    # n = len(X)
    # area = 0.0
    # j = n - 1
    # for i in range(0, n):
    #     area += (X[j] + X[i]) * (Y[j] - Y[i])
    #     j = i
    # return int(abs(area / 2.0))

def dist(x1, y1, x2, y2):
    return np.linalg.norm(np.array([x1, y1]) - np.array([x2, y2]))
```

```python
data = []

def fill(folder, y):
    for file in os.listdir(folder):
        df = pd.read_csv(os.path.join(folder, file))
        df.columns = df.columns.str.replace(' ','')
        eye_left, eye_right = [], []
        mouth_left, mouth_right = [], []
        mdistances = []
        mau = []
        c = 0
        for _, row in df.iterrows():
            if row.loc['success'] != 1.0:
                continue
            distances = []
            c += 1
            mau.append(row.loc[['AU01_r', 'AU02_r', 'AU04_r', 'AU05_r', 'AU06_r', 'AU07_r', 'AU09_r', 'AU10_r', 'AU12_r', 'AU14_r', 'AU15_r', 'AU17_r', 'AU20_r', 'AU23_r', 'AU25_r', 'AU26_r', 'AU45_r']].to_list())
            eye_left.append(calc_area(row.loc[['x_36', 'x_37', 'x_38', 'x_39', 'x_40', 'x_41']], row.loc[['y_36', 'y_37', 'y_38', 'y_39', 'y_40', 'y_41']]))
            eye_right.append(calc_area(row.loc[['x_42', 'x_43', 'x_44', 'x_45', 'x_46', 'x_47']], row.loc[['y_42', 'y_43', 'y_44', 'y_45', 'y_46', 'y_47']]))


            mouth_left.append(calc_area(row.loc[['x_48', 'x_49', 'x_50', 'x_51', 'x_57', 'x_58', 'x_59']], row.loc[['y_48', 'y_49', 'y_50', 'y_51', 'y_57', 'y_58', 'y_59']]))
            mouth_right.append(calc_area(row.loc[['x_51', 'x_52', 'x_53', 'x_54', 'x_55', 'x_56', 'x_57']], row.loc[['y_51', 'y_52', 'y_53', 'y_54', 'y_55', 'y_56', 'y_57']]))


            # dist calc.
            distances.append(dist(row.loc['x_17'],row.loc['y_17'],row.loc['x_21'],row.loc['y_21']))
            distances.append(dist(row.loc['x_22'],row.loc['y_22'],row.loc['x_26'],row.loc['y_26']))
            distances.append(dist(row.loc['x_19'],row.loc['y_19'],row.loc['x_39'],row.loc['y_39']))
            distances.append(dist(row.loc['x_24'],row.loc['y_24'],row.loc['x_42'],row.loc['y_42']))
            distances.append(dist(row.loc['x_36'],row.loc['y_36'],row.loc['x_39'],row.loc['y_39']))
            distances.append(dist(row.loc['x_42'],row.loc['y_42'],row.loc['x_45'],row.loc['y_45']))
            distances.append(dist(row.loc['x_38'],row.loc['y_38'],row.loc['x_40'],row.loc['y_40']))
            distances.append(dist(row.loc['x_43'],row.loc['y_43'],row.loc['x_47'],row.loc['y_47']))
            distances.append(dist(row.loc['x_39'],row.loc['y_39'],row.loc['x_48'],row.loc['y_48']))
            distances.append(dist(row.loc['x_42'],row.loc['y_42'],row.loc['x_54'],row.loc['y_54']))
            distances.append(dist(row.loc['x_33'],row.loc['y_33'],row.loc['x_48'],row.loc['y_48']))
            distances.append(dist(row.loc['x_33'],row.loc['y_33'],row.loc['x_54'],row.loc['y_54']))
            distances.append(dist(row.loc['x_33'],row.loc['y_33'],row.loc['x_51'],row.loc['y_51']))
            distances.append(dist(row.loc['x_51'],row.loc['y_51'],row.loc['x_48'],row.loc['y_48']))
            distances.append(dist(row.loc['x_51'],row.loc['y_51'],row.loc['x_54'],row.loc['y_54']))
            distances.append(dist(row.loc['x_48'],row.loc['y_48'],row.loc['x_54'],row.loc['y_54']))
            distances.append(dist(row.loc['x_51'],row.loc['y_51'],row.loc['x_57'],row.loc['y_57']))
            distances.append(dist(row.loc['x_48'],row.loc['y_48'],row.loc['x_57'],row.loc['y_57']))
            distances.append(dist(row.loc['x_54'],row.loc['y_54'],row.loc['x_57'],row.loc['y_57']))

            mdistances.append(distances)
        
        datav = []
        vid = np.array(mdistances)
        au = np.array(mau)
        datav.extend(list(np.mean(vid, axis = 0)))
        datav.extend(list(np.mean(au, axis = 0)))
        datav.append(np.mean(np.array(eye_left)))
        datav.append(np.mean(np.array(eye_right)))
        datav.append(np.mean(np.array(mouth_left)))
        datav.append(np.mean(np.array(mouth_right)))

        datav.extend(list(np.std(vid, axis = 0)))
        datav.extend(list(np.std(au, axis = 0)))
        datav.append(np.std(np.array(eye_left)))
        datav.append(np.std(np.array(eye_right)))
        datav.append(np.std(np.array(mouth_left)))
        datav.append(np.std(np.array(mouth_right)))

        datav.append(int(y))

        data.append(datav)


for folder in os.listdir(d):
    if folder == "Bells_csv":
        fill(os.path.join(d, folder), True)
    else:
        fill(os.path.join(d, folder), False)
```

```python
with open("input_1.csv", "w") as file:
    for row in data:
        file.write(",".join(list(map(str, row))))
        file.write("\n")
        print(row)
```
