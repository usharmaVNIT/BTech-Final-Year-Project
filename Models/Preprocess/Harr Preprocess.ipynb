```python
import cv2
import os

trained_face_data = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
```

```python
path_base = '/content/drive/MyDrive/Final year project/CNN_test/Images/train'
target_path = '/content/drive/MyDrive/Final year project/CNN_test/Cropped_Images/train'

os.makedirs(target_path)

print(os.listdir(path_base))

def get_face(path, image):
    gray_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    face_coordinates = trained_face_data.detectMultiScale(gray_img)
    try:
        y, x, h, w = face_coordinates[0]
        return True, image[x: x + w + 1, y: y + h + 1]
    except:
        print(path)
        return False, 0

i = 0
for folder in os.listdir(path_base):
    target_class_folder = os.path.join(target_path, folder)
    os.makedirs(target_class_folder)
    folder_path = os.path.join(path_base, folder)
    for video in os.listdir(folder_path):
        video_path = os.path.join(folder_path, video)
        for frame in os.listdir(video_path):
            val, face = get_face(os.path.join(video_path, frame), cv2.imread(os.path.join(video_path, frame)))
            if val:
                cv2.imwrite(os.path.join(target_class_folder, str(i) + ".jpg"), face)
                i += 1
```
